```python
import pandas as pd
from src.chunking import TextChunker
from src.embedding import Embedder
from src.indexing import ChromaIndexer

# STEP 1 — Load cleaned dataset
df = pd.read_csv("../data/processed/filtered_complaints.csv")

# STEP 2 — Chunk text
chunker = TextChunker(chunk_size=500, chunk_overlap=100)
chunk_df = chunker.chunk_dataframe(df)
```

```python
# STEP 3 — Embed chunks
embedder = Embedder()
embedded_df = embedder.embed_chunks(chunk_df)
```

```python
# STEP 4 — Index into ChromaDB
from src.indexing import ChromaIndexer  # import your custom class

# Initialize the Chroma indexer
indexer = ChromaIndexer(persist_dir="vector_store")

# Index your dataframe of chunks
# embedded_df should have columns: chunk_id, complaint_id, product, chunk_text
indexer.index(embedded_df)

print("✓ All chunks embedded and stored in ChromaDB!")
```
